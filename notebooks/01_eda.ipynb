```python
df["Class"].value_counts()
```

```python
#Fraud Percentage
fraud_percentage = df["Class"].mean() * 100
fraud_percentage
```

```python
#Visualization of Class Distribution
df["Class"].value_counts().plot(kind="bar")
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
#Transaction amounts by class
df.groupby("Class")["Amount"].describe()
```

```python
df["Amount"].hist(bins=50)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#Histogram Fraud Transaction Amounts
df[df["Class"] == 1]["Amount"].hist(bins=50)
plt.title("Fraud Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#Average Transaction Amount for fraudulent and non-fraudulent transactions
df.groupby("Class")["Amount"].mean()
```

```python
df["Time"].hist(bins=50)
plt.title("Time Distribution")
plt.xlabel("Seconds Since First Transaction")
plt.ylabel("Frequency")
plt.show()
```

```python
df.corr()["Class"].sort_values(ascending=False)
```

Key Findings
The dataset contains 284,807 transactions and 31 columns.
There are no missing values.
Fraud accounts for only about 0.17% of all transactions.
Transaction amounts are heavily right-skewed.
Several PCA-transformed features show correlation with fraud.
